```python
import pandas as pd
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
## drop unnecessary columns from the dataset
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
## perform levelencoding on Gender
from sklearn.preprocessing import LabelEncoder
lab_enc_gen=LabelEncoder()
df['Gender']=lab_enc_gen.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_geo=OneHotEncoder()
geo_transf=one_hot_geo.fit_transform(df[['Geography']])
```

```python
geo_transf.toarray()
```

```python
one_hot_geo.categories_
```

```python
data=pd.DataFrame(geo_transf.toarray(),columns=['France','Germany','Spain'])
```

```python
data.head()
```

```python
df=pd.concat([df,data],axis=1)
```

```python
df.head()
```

```python
df.drop(['Geography'],axis=1,inplace=True)
```

```python
df.info()
```

```python
## convert the datatypes
columns=['France','Germany','Spain']
for column in columns:
    df[column]=df[column].astype(int)
```

```python
df.head()
```

```python
import pickle
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(lab_enc_gen,file)
```

```python
with open('one_hot_encoder_geo.pkl','wb') as file:
    pickle.dump(one_hot_geo,file)
```
